```python
%%writefile bst.c
#include <stdio.h>
#include <stdlib.h>

// Estrutura do nó da árvore
typedef struct No {
    int valor;
    struct No* esq;
    struct No* dir;
} No;

// Estrutura da BST: guarda o ponteiro da raiz (e pode ter mais campos se desejar)
typedef struct BST {
    No* raiz;
    // Você pode incluir também campos como "int tamanho;" etc.
} BST;

// Cria um novo nó
No* criaNo(int valor) {
    No* novo = (No*)malloc(sizeof(No));
    if (novo == NULL) {
        printf("Erro ao alocar memória.\n");
        exit(1);
    }
    novo->valor = valor;
    novo->esq = NULL;
    novo->dir = NULL;
    return novo;
}

// Cria uma nova BST (inicialmente vazia)
BST* criaBST() {
    BST* bst = (BST*)malloc(sizeof(BST));
    if (bst == NULL) {
        printf("Erro ao alocar memória para a BST.\n");
        exit(1);
    }
    bst->raiz = NULL;
    return bst;
}

// Função auxiliar recursiva para inserir na BST
No* insereRec(No* raiz, int valor) {
    if (raiz == NULL) {
        return criaNo(valor);
    }
    if (valor < raiz->valor) {
        raiz->esq = insereRec(raiz->esq, valor);
    } else {
        raiz->dir = insereRec(raiz->dir, valor);
    }
    return raiz;
}

// Função pública de inserção
void insereBST(BST* bst, int valor) {
    if (bst == NULL) return;
    bst->raiz = insereRec(bst->raiz, valor);
}

// Função auxiliar recursiva de busca
No* buscaRec(No* raiz, int valor) {
    if (raiz == NULL || raiz->valor == valor) {
        return raiz;
    }
    if (valor < raiz->valor) {
        return buscaRec(raiz->esq, valor);
    } else {
        return buscaRec(raiz->dir, valor);
    }
}

// Função pública de busca
No* buscaBST(BST* bst, int valor) {
    if (bst == NULL) return NULL;
    return buscaRec(bst->raiz, valor);
}

// Percurso em-ordem recursivo
void emOrdemRec(No* raiz) {
    if (raiz != NULL) {
        emOrdemRec(raiz->esq);
        printf("%d ", raiz->valor);
        emOrdemRec(raiz->dir);
    }
}

// Função pública de percurso em-ordem
void emOrdemBST(BST* bst) {
    if (bst == NULL) return;
    emOrdemRec(bst->raiz);
}

// Função auxiliar recursiva para liberar nós
void liberaArvoreRec(No* raiz) {
    if (raiz != NULL) {
        liberaArvoreRec(raiz->esq);
        liberaArvoreRec(raiz->dir);
        free(raiz);
    }
}

// Função pública para liberar a BST
void liberaBST(BST* bst) {
    if (bst == NULL) return;
    liberaArvoreRec(bst->raiz);
    free(bst);
}

int main() {
    // Cria uma BST
    BST* arv = criaBST();

    // Insere alguns valores
    insereBST(arv, 50);
    insereBST(arv, 30);
    insereBST(arv, 70);
    insereBST(arv, 20);
    insereBST(arv, 40);
    insereBST(arv, 60);
    insereBST(arv, 80);

    // Percorre em-ordem
    printf("Valores em-ordem: ");
    emOrdemBST(arv);
    printf("\n");

    // Faz uma busca
    int valorParaBuscar = 40;
    No* encontrado = buscaBST(arv, valorParaBuscar);
    if (encontrado) {
        printf("Valor %d encontrado na árvore.\n", valorParaBuscar);
    } else {
        printf("Valor %d não encontrado.\n", valorParaBuscar);
    }

    // Libera toda a memória
    liberaBST(arv);
    return 0;
}
```
